We observe that the different digits are still well defined after binarization based on the Otsu threshold algorithm. Moreover, the digits' contours (in blue)  are also accurate enough.

```python
"""
Calculate the DFT of the digit's contour 
Return the first four coefficient
"""
def Fourier_descriptors (image) :
    # get the contours
    contour = find_contours(image,4.0)
    contour = np.squeeze(np.asarray(contour))
    # make a difference if the zeros' holes are filled or not
    # If not get a tuple of contour, we need to concatenate them into one array to calculate the DFT
    if (contour.ndim==1) :
        contour_concatenated = np.concatenate((contour[0],contour[1]), axis=0)
        contour_complex = np.empty(contour_concatenated.shape[0], dtype=complex)
        contour_complex.real = contour_concatenated[:, 1]
        contour_complex.imag = contour_concatenated[:, 0]
        DFT = np.fft.fft(contour_complex)
    else :
        contour_complex = np.empty(contour.shape[0], dtype=complex)
        contour_complex.real = contour[:, 0]
        contour_complex.imag = contour[:, 1]
        fourier_result = np.fft.fft(contour_complex)
        DFT = np.fft.fft(contour_complex)
    
    return DFT[:4]
```

```python
def plot_two_Fourier_descriptor(image1, image2 ,image1_name, image2_name, coeff1,coeff2, label1, label2 ) :
    for im,name in zip(image1,image1_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        zero = plt.scatter(A1,A2,c='red')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    for im, name in zip (image2,image2_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        one = plt.scatter(A1,A2,c='blue')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    plt.xlabel(label1)
    plt.ylabel(label2);
    plt.legend((zero,one),('zeros','ones'))
```

Therefore, the coefficient A1 is a good feature to linearly separate the digits and discriminate them later on as all zeros' amplitudes are lower than the ones'. The (A0,A1) combination may be linearly separable as well, but the 0.3 and 1.8 digits are very close in A1 and in general as well, they surely are separable with a slightly more complicated boundary than a linear one. In conclusion, we can separate the ones and zeros using this simplified representation. The best combination is achieved with (A1,A3), because the variance in the cluster is the smallest.

#### Invariance properties
In this subsection we describe the properties of the above descriptions. The fourier descriptors allows to create descriptions which could be translation, rotation and/or scaling invariant depending on the application. This depends on the choice of the features based on the coefficient of the DFT. Here we only asses the rotation invariance.

Note that using the amplitude of the DFT coefficient gets rid of all effects that affects the phase such as rotation and the choice of the starting point of the contour. Therefore, all our descriptions above are invariant to both.

A translation only affects the first coefficient, so one needs to exclude it to build a translation invariant description.

The scaling affects each coefficient by a constant factor, therefore one needs to take the ratio of the DFT coefficient to build a scaling invariant description.

##### Rotation invariance

```python
def rotate_90_images(image) :
    rotated_images = np.empty(image.shape)
    for im, idx in zip(image,range(image.shape[0])) :
        rotated_images[idx] = rotate(image[idx],90)
    return rotated_images
```

On the graph above we plot two combinations of Fourier descriptors for the original binarized and filled set and a rotated version. Comparing each original plot (left) to its rotated version (right), we can see they are pretty similar, even if a difference exists (especially for digit 1.9 in the (A1,A2) plot). This difference should not come from the rotation with the function from skimage. In fact, this function rotates the image about its center and not the center of the object (they are very similar though), thus we have a translation and rotation. However, we choose carefully not to include A0 which is the only descriptor altered by a translation. Moreover, we rotate by 90 degrees to avoid interpolation errors due to resizing and other effects. It probably comes from the interpolation from the function find_contour which allows us to get the pixel indices of the contour. An approach to confirm that will be to take the contour we get with `find_contour` without rotation of the image and apply a rotation on the contour only and then calculate the DFT. Here we do not do that and we conclude that it is rotation invariant.

### Region-based descriptors

The first step was to do some preprocessing steps on the images. We started by binarizing the images using the Otsu threshold method and then we performed an opening to remove white pixels that were not part of the digit.  

```python
def binary_im(im):
    # Binarize the image
    thresh = threshold_otsu(im)
    return im > thresh
```

After this preprocessing, we computed several region-based descriptors in order to discriminate our digits in a 2D plane. The descriptors studied in this part are: 

- area: the number of pixels of the region
- perimeter: approximation of the contour as a line through the centers of border pixels using a 4-connectivity
- compacity: the ratio of squared perimeter over area
- convex area: number of pixels of the smallest convex polygon that encloses the region
- filled area: number of pixels of the region with holes filled in
- horizontal projection: sum all the pixels values in each column (this results in the bar plot below) and compute the width of the projection (= compute how many non-zeros components per projection)  
- elongation: square root of the ratio of the eigenvalues of the matrix of inertia (or covariance matrix)
- horizontal variance: first element of the diagonal of the covariance matrix 
- vertical variance: second element of the diagonal of the covariance matrix 
- moments: spatial moments up to the 3rd order 
- gravity center: its coordinates are given by ratios of moments (M1,0/M0,0 , M0,1/M0,0) as defined in the course
    
Then we plotted those descriptors against each other in order to see whether they could help discriminate our digits. Another important step before doing these calcalutations was to verify that the labeled object on each image was really the digit and not some random "lost" white pixels. To do so, the function `biggest_label` assigns as label the biggest labeled object on the image.  

```python
def biggest_label(properties):
    # assigns to max_lab the biggest labeled object on the image
    max_lab = properties[0]
    for lab in properties:
        if lab.area > max_lab.area:
            max_lab = lab
    return max_lab
```

```python
def region_based_descr(bin_im):
    # compute all region-based descriptors 
    labels = label(bin_im)
    biggest_lab = biggest_label(regionprops(labels))
    area = biggest_lab.area
    perimeter = biggest_lab.perimeter
    compacity = perimeter**2 / area 
    convex_area = biggest_lab.convex_area
    filled_area = biggest_lab.filled_area
    h_projection = np.count_nonzero(sum(bin_im, 0))
    elongation = math.sqrt(inertia_tensor_eigvals(bin_im)[0]/inertia_tensor_eigvals(bin_im)[1])
    h_variance = inertia_tensor(bin_im)[0,0]
    v_variance = inertia_tensor(bin_im)[1,1]
    moment = moments(bin_im)
    gravCenter = (moments(bin_im)[1, 0] / moments(bin_im)[0, 0], moments(bin_im)[0, 1] / moments(bin_im)[0, 0])
    
    return compacity, area, convex_area, perimeter, filled_area, h_projection, elongation, h_variance, v_variance, moment, gravCenter
```

As expected, most of the region-based descriptors allow to discriminate the two digits as they have quite a different shape and therefore different geometrical properties. 

For the area, we expected quite different results between the ones and the zeros as the zeros seem to visually have a bigger area. This is emphasized when using filled or convex area. As can be seen on the images, the perimeter of a zero is more important than for the ones. This is confirmed on the plots above. Moreover, as the compacity is a ratio between the squared perimeter and the area, it therefore also follows the same tendancy.  

Moreover, for the horizontal projection, we expected a larger value for the zeros as they are larger than the ones, meaning that the bar plot will be broader in the case of zeros. This is confirmed on the bar plots and on the plots above. 

Concerning the elongation, as said before it is the squared root of the ratio of the eigenvalues of the matrix of inertia. Moreover, as stated in the lecture the eigenvalues represent the variance of the shape projected on the axes of inertia. Therefore, once again, the variance of the zeros will be higher due to their shape. 

As for the horizontal and vertical variances, when observing the digits it was clearly visible that the horizontal variance would be more important for the zeros, but that the vertical variance would be approximately in the same range for both digits, and this can be seen on the plot horizontal variance VS vertical variance. 

Unfortunately, the gravity center was not really useful to discrimate both digits. Indeed, the are all in a range from 13 to 15. On the plot, we see that all points are mixed and do not form any cluster.

Finally, moments were quite good at separating the ones from the zeros as can be seen on the last two plots above. As stated in the lecture, the moments are used in physics to describe the mass repartition of a body, and as the zeros and the ones have qui a different shape, they should have different mass repartition and indeed their moments have quite different values.

```python
def plot_two_Fourier_descriptor_4digits(image1, image2,image3, image4,image1_name, image2_name,image3_name, image4_name, 
                                        coeff1,coeff2, label1, label2 ) :
    for im,name in zip(image1,image1_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        zero = plt.scatter(A1,A2,c='red')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    for im, name in zip (image2,image2_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        one = plt.scatter(A1,A2,c='blue')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    for im,name in zip(image3,image3_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        two = plt.scatter(A1,A2,c='green')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    for im, name in zip (image4, image4_name):
        coeff = Fourier_descriptors(im)
        A1 = np.sqrt(coeff[coeff1].real**2+coeff[coeff1].imag**2)
        A2 = np.sqrt(coeff[coeff2].real**2+coeff[coeff2].imag**2)
        three = plt.scatter(A1,A2,c='black')
        plt.text(A1+0.3, A2+0.3, name, fontsize=9)
    plt.xlabel(label1)
    plt.ylabel(label2);
    plt.legend((zero,one,two,three),('zeros','ones','twos','threes'))
```
