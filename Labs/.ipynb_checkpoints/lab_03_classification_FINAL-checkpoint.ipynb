First we observe that the three classes' distribution are nearly ellipsoidal. We can see that classes a and b have the same dispersion around different center in the plot. On the other hand, classes c's dispersion is different. If we were to model the data as three classes following normal laws, we would have similar covariance matrices for a and b, but a different one for c. Moreover, they would all have different means.

#### Normal pdf assumption and parameters estimation
Here we assume the three classes to follow a two-dimensional gaussian law: $a,b,c \sim \mathcal{N}(\mu\,\Sigma)\$ 

We estimates $\mu$ and $\Sigma$ of each classes using maximum likelihood estimators : 

$\hat{\mu} :=\frac{1}{N}\sum_{i = 1}^{N}\ x_{i}$

$\hat{\Sigma} :=\frac{1}{N}\sum_{i = 1}^{N}\ (x_{i} -\hat{\mu})(x_{i} -\hat{\mu})^{T}$

```python
def get_parameters(X) :
    """
        return :
        1) the maximum likelihood estimator of the mean
        2) the maximum likelihood estimator of the covariance matrix
    """
    X_mean = np.mean(X, axis =0)
    X_centered = X - X_mean
    X_cov = (1./X.shape[0])*X_centered.T.dot (X_centered)
    return np.rint(X_mean), np.rint(X_cov)
```

```python
### Estimate the parameters 
a_mean, a_cov = get_parameters(a)
b_mean, b_cov = get_parameters(b)
c_mean, c_cov = get_parameters(c)

print('Class a : \n mean :', a_mean,"\n covariance matrix:\n ",a_cov)
print('\n Class b :\n mean :', b_mean,"\n covariance matrix:\n",b_cov)
print('\n Class c :\n mean :', c_mean,"\n covariance matrix:\n ",c_cov)
```

As expected classes a and b have the same covariance matrix with a different mean and class c has a different covariance matrix compared to the two other classes. Moreover, all three classes have a $\hat{\Sigma}_{MLE} $ which is diagonal, meaning that the two dimensions are uncorrelated. 

#### Bayesian Classification for normal laws

Bayesian classifier's decision surface between  classes i and j is given by : 

$$P(w_{i}|x) - P(w_{j}|x) = 0\quad\text{(1)}$$ where : $$P(w_{i}|x)\quad\text{is the posterior probability of class (i)      (2)}$$ 

We basically assign the most probable class of the two by finding the boundary where they are equiprobable, thus minimizing the probability of error.

We can also write (2) as :  $$g_{i} = f(P(w_{i}|x))\quad\text{(3)}$$  

where f() is a monotonally increasing function called discriminant function.

Rewriting (1) with (3) gives :

$$ g_{ij}(x) = g_{i}(x) - g_{j}(x) = 0\quad\quad\text{(4)}$$

If we make the assumption that our classes are gaussian :  $a,b,c \sim \mathcal{N}(\mu\,\Sigma)\$ and estimates the respective parameters, we can write the explicit formula of (3) using ln(x) as f(x) :

$$g_{i}(x) = - \frac{1}{2}(x - μ_{i})^{T}Σ_{i}^{-1}(x - μ_{i}) + ln(P(w_{i})) + ln(\frac{1}{2\pi\sqrt{|Σ_{i}|}})\quad\quad\text{(5)}$$

where: $$ P(w_{i}) \quad\text {is the prior probability of class (i)  (6)}$$ 

$$ \frac{ X\in i}{X\in a,b,c} = 200/600 = 1/3, i\in{1,2,3} \quad\text {the estimated prior probability of class (i) }$$

Here we assumed that the samples provided are a good representation of the ground truth and we have enough data to obtain accurate estimators of the priors.

##### Decision curve between class a and b
We have  $ \Sigma_{MLE,a}= \Sigma_{MLE,b}$, therefore rewriting (5) above we get :

$$g_{i}(x) = w_{i}^{T}x + w_{i0}\quad\quad\text{(1)}$$
where:
$$w_{i} = Σ^{-1}μ_{i}$$
$$w_{i0} = ln(P(w_{i}))- \frac{1}{2}μ_{i}^{T}Σ^{-1}μ_{i}$$

and 

$$g_{ij}(x) = g_{i}(x) - g_{j}(x) =  (w_{i}^{T}-w_{j}^{T})x + w_{i0}-w_{j0}\quad\quad\text{(2)}$$

We can see that the decision surface is linear in x, hence the decision surface is a hyperplane. Moreover, the priors are equal for all classes and will disappear from the constant term.

##### Decision curves between class a and c , b and c
Here $ \Sigma_{MLE,a} or \Sigma_{MLE,b} \neq \Sigma_{MLE,c}$, therefore there is no simplification of the discriminant function : $$g_{i}(x) = - \frac{1}{2}(x - μ_{i})^{T}Σ_{i}^{-1}(x - μ_{i}) + ln(P(w_{i})) + ln(\frac{1}{2\pi\sqrt{|Σ_{i}|}})\quad\quad\text{(1)}$$ 

The decision boundary is given by :
$$ g_{ij}(x) = g_{i}(x) - g_{j}(x) = 0\quad\quad\text{(2)}$$

Developing the expression gives us :

$$g_{i}(x) - g_{j}(x) = -\frac{1}{2}x^{T}Σ_{i}^{-1}x +\frac{1}{2}x^{T}Σ_{j}^{-1}x + ln(P(w_{i})) + ln(\frac{1}{2\pi\sqrt{|Σ_{i}|}})-ln(P(w_{j})) - ln(\frac{1}{2\pi\sqrt{|Σ_{j}|}})\quad\quad\text{}(3)$$

Using the fact that we have equal priors, we can simplify the above expresion:

$$g_{i}(x) - g_{j}(x) = -\frac{1}{2}x^{T}Σ_{i}^{-1}x +\frac{1}{2}x^{T}Σ_{j}^{-1}x + ln(\frac{1}{2\pi\sqrt{|Σ_{i}|}})- ln(\frac{1}{2\pi\sqrt{|Σ_{j}|}})\quad\quad\text{}(3)$$

The decision boundary is quadratic in x, so the decision curve is a hyperquadric.

```python
def get_hyperplane(mean_1,mean_2,cov_1,cov_2, X1) :
    """
    Take a vector of X1 coordinate and return the associated X2 coordinate corresponding to the decision boundary
    """
    cov_1_inv = np.linalg.inv(cov_1)
    cov_2_inv =np.linalg.inv(cov_2)
    weights = np.dot(cov_1_inv,mean_1.T)-np.dot(cov_2_inv,mean_2.T).T
    constant = 0.5*((np.dot(mean_2,np.dot(cov_2_inv,mean_2.T)))-(np.dot(mean_1,np.dot(cov_1_inv,mean_1.T))))
    hyperplane = -(weights[0]*(X1) + constant)/weights[1]
    return hyperplane
```

```python
def get_hyperquadric(x,y, mean_1,mean_2, cov_1,cov_2):
    """
    Take a meshgrid of points (X1,X2) and compute the discriminant for that value
    Return a matrix of discriminant
    Use the zero level to plot the decision boundary 
    """
    ### Discriminant 1 
    
    # create a new axis : (2,) to (2,1)
    mean_1 = np.reshape(mean_1,(2,1))
    # create a (200,200,2,1) where (2,1) are the corrdinate (x,y)
    x = np.stack((x,y), axis=-1)[..., np.newaxis]
    # create a a covariance matrix of shape (1,1,2,2)
    cov_1_inv  = np.reshape(np.linalg.inv(cov_1),(1,1,2,2))
    # create a a mean vector of shape (1,1,1,2)
    mean_1_T = np.reshape(mean_1.transpose(),(1,1,1,2))
    c1 = np.log(1/(2*np.pi * np.sqrt(np.linalg.det(cov_1))))
    # swap the last two axis for the matrix calculus
    x_swap = np.moveaxis(x, -1, -2)
    quadratic1  =  -(1/2)* x_swap  @ cov_1_inv @ x
    linear1 = 1/2 * x_swap @ cov_1_inv @ mean_1 + 1/2 * mean_1_T @ cov_1_inv @ x
    c_1 = 1/2 * mean_1_T @ cov_1_inv @ mean_1
    
    # get rid of the last two axis : (200,200,1,1) to (200,200)
    g1 = (quadratic1 + linear1 - c_1 + c1)[...,0,0]
    
    ### Discriminant 2 
    
    # create a new axis : (2,) to (2,1)
    mean_2 = np.reshape(mean_2,(2,1))
    # create a a covariance matrix of shape (1,1,2,2)
    cov_2_inv  = np.reshape(np.linalg.inv(cov_2),(1,1,2,2))
    # create a a mean vector of shape (1,1,1,2)
    mean_2_T = np.reshape(mean_2.transpose(),(1,1,1,2))
    c2 = np.log(1/(2*np.pi * np.sqrt(np.linalg.det(cov_2))))
    quadratic2 =  -(1/2)* x_swap @ cov_2_inv @ x
    linear2 = 1/2 * x_swap @ cov_2_inv @ mean_2 + 1/2 * mean_2_T @ cov_2_inv @ x
    c_2 = 1/2 * mean_2_T @ cov_2_inv @ mean_2
    
    # get rid of the last two axis : (200,200,1,1) to (200,200)
    g2 = (quadratic2 + linear2 - c_2 + c2)[...,0,0]
    
    return g1-g2
```

```python
def compute_error_binary (mean_1,mean_2,cov_1,cov_2, X, plane) :
    """
    Take a vector of points corresponding to the concatenation of two of the training classes
    Need to know if the decision boundary is a hyperplane or not to compute the discriminant
    Classify the points according to the discriminant
    Return the two vectors corresponding to the classification to the two classes and the training error made
    return the vector of discriminant as well
    """
    # Check if region boundary is a plane
    if(plane == True) :
        cov_1_inv = np.linalg.inv(cov_1)
        cov_2_inv =np.linalg.inv(cov_2)
        weights = np.dot(cov_1_inv,mean_1.T)-np.dot(cov_2_inv,mean_2.T).T
        constant = 0.5*((np.dot(mean_2,np.dot(cov_2_inv,mean_2.T)))-(np.dot(mean_1,np.dot(cov_1_inv,mean_1.T))))
        # calculate the value of the discriminant for the vector of point X
        discriminant = np.dot(weights,X.T) + constant
    # check if the decision boundary is a hyperquadratic
    if(plane==False) :
        cov_1_inv = np.linalg.inv(cov_1)
        cov_2_inv =np.linalg.inv(cov_2)
        # Same principle as in get_hyperquadratic, we reshape the matrices for the matmul operation
        X = np.reshape(X,(X.shape[0],1,2))
        X_swap = np.reshape(X,(X.shape[0],2,1))
        mean_1 = np.reshape(mean_1,(2,1))
        mean_2 = np.reshape(mean_2,(2,1))
        mean_1_T = np.reshape(mean_1.transpose(),(1,1,2))
        mean_2_T = np.reshape(mean_2.transpose(),(1,1,2))
        quadratic1  =  -(1/2)* X  @ cov_1_inv @ X_swap
        linear1 = 1/2 * X @ cov_1_inv @ mean_1 + 1/2 * mean_1_T @ cov_1_inv @ X_swap
        c_1 = 1/2 * mean_1_T @ cov_1_inv @ mean_1
        c1 = np.log(1/(2*np.pi * np.sqrt(np.linalg.det(cov_1))))
        quadratic2 =  -(1/2)* X @ cov_2_inv @ X_swap
        linear2 = 1/2 * X @ cov_2_inv @ mean_2 + 1/2 * mean_2_T @ cov_2_inv @ X_swap
        c_2 = 1/2 * mean_2_T @ cov_2_inv @ mean_2
        c2 = np.log(1/(2*np.pi * np.sqrt(np.linalg.det(cov_2))))
        g1 = (quadratic1 + linear1 - c_1 + c1)[...,0,0]
        g2 = (quadratic2 + linear2 - c_2 + c2)[...,0,0]
        # Compute the discriminant for the vector of point X
        discriminant =g1-g2
        
    # get the indices of the discriminant where it is greater than 0, it will be classified as C1
    C1 = np.argwhere(discriminant > 0)
    # indices greater than 200 are misclassified as the first 200 point are from C1
    missclassified_c1 = np.sum(C1 >= 200,axis=0)
    # get the indices of the discriminant where it is smaller than 0
    C2= np.argwhere(discriminant<0)
    # indices lower than 200 are misclassified as the last 200 point are from C2
    missclassified_c2 = np.sum(C2 < 200,axis=0)
    #Calculate the misclassification error
    error = (missclassified_c1[0] + missclassified_c2[0])/X.shape[0]
    
    return X[C1],X[C2], error*100,discriminant
```

```python
def get_boundaries(a_mean,b_mean,c_mean,a_cov,b_cov,c_cov) :
    """
    Return all the decision boundaries : ab,ac,bc
    """
    # hyperplane ab
    x_ab = np.linspace(-15,15, 1000)
    y_ab = get_hyperplane(a_mean,b_mean,a_cov,b_cov, x_ab)

    ##hyperquadric 
    x_quadratic = np.arange(-10, 10, 0.1)
    y_quadratic = np.arange(-10, 10, 0.1)
    x, y = np.meshgrid(x_quadratic, y_quadratic)
    #ac
    Z_ac = get_hyperquadric(x,y, a_mean,c_mean, a_cov,c_cov)
    #bc
    Z_bc = get_hyperquadric(x,y, b_mean,c_mean, b_cov,c_cov)
    
    return x,y,x_ab,y_ab,Z_ac,Z_bc
```

```python
def plot_binary (data1,data2,label1,label2,x,y,x_ab,y_ab,Z_ac,Z_bc,plane) :
    """
    Plot data1 and data2 on a scatter plot
    Plot the decision boundaries : Use the zero level set of the discriminant matrix to plot the hyperquadric
    Plot the the a hyperplane if plane equal True or hyperquadric if False
    """
    if (plane == True) :
        plt.scatter(data1[:,0], data1[:, 1], label=label1)
        plt.scatter(data2[:,0], data2[:, 1], label=label2)
        plt.plot(x_ab, y_ab,color = 'red',label='ab boundary')
        plt.legend()
        plt.grid(True)
        plt.xlabel('X1')
        plt.ylabel('X2')
        plt.xlim(-15.0,15.0)
        plt.ylim(-10.0,10.0)
    if(plane == False) :
        plt.scatter(data1[:,0], data1[:, 1], label=label1)
        plt.scatter(data2[:,0], data2[:, 1], label=label2)
        if( label1 == 'a') :
            ac_boundary = plt.contour(x, y, Z_ac,colors ='k', levels = 0)
            ac_boundary.collections[0].set_label('ac boundary')
        if (label1 == 'b') :
            bc_boundary = plt.contour(x, y, Z_bc,colors= 'purple', levels = 0)
            bc_boundary.collections[0].set_label('bc boundary')
        plt.legend()
        plt.grid(True)
        plt.xlabel('X1')
        plt.ylabel('X2')
        plt.xlim(-15.0,15.0)
        plt.ylim(-10.0,10.0)
```

```python
def plot_three_classes(data1,data2,data3,x,y,x_ab,y_ab,Z_ac,Z_bc) :
    """
    Plot data1, data2 and data3 on a scatterplot
    Plot the region boundaries : use the zero level set of the discriminant matrix for the hyperquadric
    """
    
    ac_boundary = plt.contour(x, y, Z_ac,colors ='k', levels = 0)
    ac_boundary.collections[0].set_label('ac boundary')
    bc_boundary = plt.contour(x, y, Z_bc,colors= 'purple', levels = 0)
    bc_boundary.collections[0].set_label('bc boundary')
    plt.scatter(data1[:,0], data1[:, 1], label='a')
    plt.scatter(data2[:,0], data2[:, 1], label='b')
    plt.scatter(data3[:,0], data3[:, 1], label='c')
    plt.plot(x_ab, y_ab,color = 'red',label='ab boundary')
    plt.grid(True)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
```

```python
# compute the decision boundaries
x,y,x_ab,y_ab,Z_ac,Z_bc = get_boundaries(a_mean,b_mean,c_mean,a_cov,b_cov,c_cov)

plt.figure(figsize=(25,25))
plt.subplot(321)
plot_three_classes(a,b,c,x,y,x_ab,y_ab,Z_ac,Z_bc)
plt.title('Bayesian classification for normal laws')
########################################################

#compute the error classification between a and b
X_ab = np.concatenate((a,b),axis=0)
a_classified,b_classified,error_ab,_ = compute_error_binary (a_mean,b_mean,a_cov,b_cov, X_ab,True)
# get rid of useless dimension which equal 1
a_classified = np.squeeze(a_classified)
b_classified = np.squeeze(b_classified)

plt.subplot(322)
plot_binary (a_classified,b_classified,'a','b',x,y,x_ab,y_ab,Z_ac,Z_bc,True)
plt.title('Binary classification of class a and b according to the discriminant function')

#compute the error classification between a and c
X_ac = np.concatenate((a,c),axis=0)
a_classified,c_classified,error_ac,_ = compute_error_binary (a_mean,c_mean,a_cov,c_cov, X_ac,False)
# get rid of useless dimension which equal 1
a_classified = np.squeeze(a_classified)
c_classified = np.squeeze(c_classified)

plt.subplot(323)
plot_binary (a_classified,c_classified,'a','c',x,y,x_ab,y_ab,Z_ac,Z_bc,False)
plt.title('Binary classification of class a and c according to the discriminant function')

#compute the error classification between b and c
X_bc = np.concatenate((b,c),axis=0)
b_classified,c_classified,error_bc,_ = compute_error_binary (b_mean,c_mean,b_cov,c_cov, X_bc,False)
# get rid of useless dimension which equal 1
b_classified = np.squeeze(b_classified)
c_classified = np.squeeze(c_classified)

plt.subplot(324)
plot_binary (b_classified,c_classified,'b','c',x,y,x_ab,y_ab,Z_ac,Z_bc,False)
plt.title('Binary classification of class b and c according to the discriminant function')

plt.show()

print('The misclassification error rate of between clusters a and b is {error_ab}%.'.format(error_ab = error_ab))
print('The misclassification error rate of between clusters a and c  is {error_ac}%.'.format(error_ac = error_ac))
print('The misclassification error rate of between clusters b and c is {error_bc}%.'.format(error_bc = error_bc))
```

The upper left panel shows all three classes and the binary decision boundaries computed using bayesian classification under normal law assumption. The three others plots perform the binary classification using those same decision boundaries and compute the misclassification error.

The ab boundary (upper right panel) is a hyperplane as expected and it classifies the data with a misclassification rate of 4.75%, because the data is not separable. 

The two other boundary are hyperquadric as already mentionned above. In the ac case the classes are not completely separable, it can be seen on the ab classification plot. Indeed, the furthest point on the left belonging to a is hidden by c points on the graph with the three classes. The misclassification rate is 0.5 % meaning that a second point is misclassified (2/400 = 0.5%), it is the blue point on the left of the black curve on the first graph. On the other hand, b and c are separable and the classification is perfect 0% misclassified points.

In order to have a classifier with three classes, we need to compute the probability of each data point to belong to each classes and assign the most probable one. In our case each probability is computed using the Gaussian assumption and the estimated parameters. To visualize it we use multiple binary classification, gaussian mixture model which does approximatively the same modelling as we did above and the distance classifier in the second part.

#### Multiple binary classification
Here we classify a point by computing the three binary classifiers using the discriminant function and we assign the class which is the most probable. To do so we look at the three comparisons ab, ac and bc and assign the class which is assigned twice (eg. ab assign a and ac assign b or ab assign b and bc assign b).

```python
def compute_error_multiple_binary(mean_a,mean_b,mean_c,cov_a,cov_b,cov_c, X):
    """
    Compute the discriminant for each of the three binary comparison for a point
    Compare the value of the discriminant for a single point and figure out which one is the most probable 
    Return the data classified and the misclassification error
    """
    
    # get the discriminant between a and b
    _,_,_,discriminant_ab = compute_error_binary(mean_a,mean_b,cov_a,cov_b, X, True)
    discriminant_ab = discriminant_ab > 0
    # get the discriminant between a and c
    _,_,_,discriminant_ac = compute_error_binary(mean_a,mean_c,cov_a,cov_c, X, False)
    discriminant_ac = discriminant_ac > 0
    # get the discriminant between b and c
    _,_,_,discriminant_bc = compute_error_binary(mean_b,mean_c,cov_b,cov_c, X, False)
    discriminant_bc = discriminant_bc > 0
    
    # Need one dimension to concatenate in axis 1
    discriminant_ab = np.reshape(discriminant_ab,(600,1))
    discriminant_ac = np.reshape(discriminant_ac,(600,1))
    discriminant_bc = np.reshape(discriminant_bc,(600,1))
    
    # matrix of discriminants each row contain the three comparison for a data point
    discriminants = np.concatenate((discriminant_ab,discriminant_ac,discriminant_bc), axis = 1)
    X_classified = np.empty((discriminants.shape[0]))
    nb_error = 0
    for idx,d in enumerate(discriminants):
        #print(idx,nb_error)
        # check if comparison ab and ac are true (>0) meaning that class a is most probable than b and c
        if(d[0] == True and d[1] == True) :
            X_classified[idx] = 0
            if (idx>199) :
                nb_error += 1
        # check if comparison ac and bc are false (<0) meaning that class c is most probable than a and b
        if (d[1] == False and d[2] == False ) :
            X_classified[idx] = 2
            if (idx<400) :
                nb_error += 1
        # check if comparison ab is false (<0) and bc is true (>0) meaning that class b is most probable than a and c
        if(d[0] == False and d[2] == True) :
            X_classified[idx] = 1
            if (idx< 200 or idx>399) :
                nb_error += 1
                
    C1 = np.argwhere(X_classified == 0)
    C2 = np.argwhere(X_classified == 1)
    C3 = np.argwhere(X_classified == 2)
    error = (nb_error / 600.0)
    
    return X[C1],X[C2],X[C3], error*100,nb_error
```

```python
# Concatenate the three classes
X_abc = np.concatenate((a,b,c),axis=0)
# Classify the data point according to the discriminant comparison
a_classified,b_classified,c_classified,error,nb_error = compute_error_multiple_binary(a_mean,b_mean,c_mean,a_cov,b_cov,c_cov, X_abc)
a_classified = np.squeeze(a_classified)
b_classified = np.squeeze(b_classified)
c_classified = np.squeeze(c_classified)
#Plot the results
plt.figure(figsize = (12,8))
plot_three_classes(a_classified,b_classified,c_classified,x,y,x_ab,y_ab,Z_ac,Z_bc)
plt.title("Classification using multiple binary comparison")
plt.show()

print('The number of erros using multiple binary classification comparison is', nb_error, 'leading to an error rate of {:.2f}%.'.format(error)) 
```

Not surprisingly, the classification matches the decision boundary as we used it to do so. Each comparison compares the most probable of the two class under Gaussian assumption and by doing multiple comparisons we take the most probable of the three classes. The global number of errors is equal to the number of errors of each individual binary classifier 19/200 = 4.75% and 2/200 = 0.5% then 19+2/600 = 3.5%, so the global number of errors is 21.

#### Gaussian Mixture Models (EM algorithm)

Inspired from : https://jakevdp.github.io/PythonDataScienceHandbook/05.12-gaussian-mixtures.html

```python
from matplotlib.patches import Ellipse
from sklearn import mixture

def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()
    
    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)
    
    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))
        
def plot_gmm(gmm, X, label=True, ax=None):
    ax = ax or plt.gca()
    # predict the class with the highest probability
    labels = gmm.fit(X).predict(X)
    # separe the labels
    a = np.argwhere(labels == 2)
    b = np.argwhere(labels == 1)
    c = np.argwhere(labels == 0)
    A = X[a].squeeze()
    B = X[b].squeeze()
    C = X[c].squeeze()
        
    if label:
        ax.scatter(A[:, 0], A[:, 1],label = 'a', color ='blue', s=40, zorder=2)
        ax.scatter(B[:, 0], B[:, 1],label = 'b', color ='orange', s=40, zorder=2)
        ax.scatter(C[:, 0], C[:, 1],label = 'c', color ='green', s=40, zorder=2)
        ax.legend()
    else:
        ax.scatter(X[:, 0], X[:, 1], s=40, zorder=2)
        
    ax.axis('equal')
    #Draw the ellipses
    w_factor = 0.2 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        draw_ellipse(pos, covar, alpha=w * w_factor)
```

```python
# concatenate the  datasets into the final training set
X_train = np.vstack([a,b,c])
# Gaussian mixture model
gmm = mixture.GaussianMixture(n_components=3, covariance_type='full', random_state=42)
# Plot the result where points are clssified to the cluster with highest probability
plt.figure(figsize=(12,8))
plot_gmm(gmm, X_train)
ac_boundary = plt.contour(x, y, Z_ac,colors ='k', levels = 0)
ac_boundary.collections[0].set_label('ac boundary')
bc_boundary = plt.contour(x, y, Z_bc,colors= 'purple', levels = 0)
bc_boundary.collections[0].set_label('bc boundary')
plt.plot(x_ab, y_ab,color = 'red',label='ab boundary')
plt.legend()
plt.title("Gaussian Mixture Models for cluster a,b and c")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
#Get the covariances and the means of the fitted Gaussian
Covs = np.rint(gmm.covariances_)
means = np.rint(gmm.means_)
print('Class a : \n mean :', means[2],"\n covariance matrix:\n ",Covs[2])
print('\n Class b :\n mean :', means[1],"\n covariance matrix:\n",Covs[1])
print('\n Class c :\n mean :', means[0],"\n covariance matrix:\n ",Covs[0])
```

The Gaussian mixture model fits three Gaussians for each classes and we can see that rounding the estimated parameters gives us the same result as our previous estimators. However, the values used for prediction are not rounded and can explain the slight difference between the classification boundaries and the actual classification of the points, if we are looking at binary classification. Here, for each data point the function computes the probability of belonging to a class according to the underlying density function and we choose to classify the points by taking the most probable class. We do not compute the error of classification as it will slightly differ comparing to the first method. We can observe that the results are similar to what we obtained with the multiple binary classification. Therefore, it verifies that the first method assigns the most probable class under a Gaussian density. 

### 1.2 Mahalanobis distance
For classes `a` and `b`, give the expression of the Mahalanobis distance used to classify a point in class `a` or `b`, and verify the obtained classification, in comparison with the "complete" Bayes classification, for a few points of the plane.

In this part we want to do a minimal distance classifier using Malahanobis distance, given by:

$$ d_m = ((x - \mu_{i})^{T}\Sigma^{-1}(x - \mu_{i}))^{1/2} $$

with x a datapoint, $\mu_{i}$ the mean and $\Sigma^{-1}$ the inverse of the covariance matrix. As can be seen in the formula, $\Sigma$ does not have an indice i because it is identical for all classes, and it has been verified in the previous section that $\Sigma_a$ = $\Sigma_b$. On the contrary, $\mu_i$ is different for each class.  

In this case, the classification is performed as follows: take a datapoint x, compute its Mahalanobis distance to $\mu_a$ and to $\mu_b$, then compare the obtained distances and the smallest one will indicate the class of x. 

```python
from scipy.spatial import distance

def mahalanobis_classifier(X):
    '''
    performs Mahalanobis classification as explained above
    returns the predicted classes of each point of X
    '''
    pred_classes = []
    for i in range(len(X)):
        dist_a = distance.mahalanobis(X[i], a_mean, np.linalg.inv(a_cov))
        dist_b = distance.mahalanobis(X[i], b_mean, np.linalg.inv(b_cov))
        
        if dist_a < dist_b:
            # if dist_a < dist_b then the class of X[i] is a -> 0
            pred_classes.append(0)
        else:
            # the class of X[i] is b -> 1
            pred_classes.append(1)
            
    return pred_classes
```

```python
# concatenate the datasets a and b 
X = np.concatenate((a, b), axis=0)

# perform classification
pred_classes = mahalanobis_classifier(X)

# create an array with the true classes of X (= of the datasets a and b)
a_classes = np.zeros(a.shape[0])
b_classes = np.ones(b.shape[0])
true_classes = np.concatenate((a_classes,b_classes), axis=None)
```

```python
# create vectors of datapoints for each predicted classes for the plot
a_pred = X[np.where(np.array(pred_classes) == 0)]
b_pred = X[np.where(np.array(pred_classes) == 1)]

# plot classification results
plt.figure(figsize=(12,7))
plot_binary (a_pred,b_pred,'a','b',x,y,x_ab,y_ab,Z_ac,Z_bc,True)
plt.title('Mahalanobis points classification and hyperplane of "complete" Bayes classification')
plt.show()

# compute the number of misclassified samples and the error rate
errors = 0
for i in range(400):
    if (pred_classes[i] != int(true_classes[i])):
        errors += 1

error_rate = (errors/(a.shape[0]+b.shape[0]))*100
print('The total number of misclassified samples using Mahalanobis distance is', errors, 'leading to an error rate of', error_rate, '%.') 
```

As can be seen above, the error rate obtained with Mahalanobis classification is 4.75%, i.e. 19 of 400 datapoints are misclassified. A first observation is that this result matched what was obtained with the binary Bayes classification above. Moreover, the decision boundary that one could imagine by just seeing the scatter plot obtained by the Mahalanobis classification corresponds  to the hyperplane obtained by the Binary  Bayes classification in red on the plot above. Hence, the two methods are equivalent. This is expected as taking the smallest distance corresponds to taking the class with the highest probability using normal law.

Now, we perform the same classification but this time with the 3 classes `a`, `b` and `c` using the same procedure, to verify our conclusions.

```python
def mahalanobis_classifier_all(X):
    '''
    performs Mahalanobis classification as explained above
    returns the predicted classes of each point of X
    '''
    pred_classes = []
    for i in range(len(X)):
        dist = []
        dist_a = distance.mahalanobis(X[i], a_mean, np.linalg.inv(a_cov))
        dist_b = distance.mahalanobis(X[i], b_mean, np.linalg.inv(b_cov))
        dist_c = distance.mahalanobis(X[i], c_mean, np.linalg.inv(c_cov))
        
        dist.append(dist_a)
        dist.append(dist_b)
        dist.append(dist_c)
        
        if min(dist) == dist_a:
            pred_classes.append(0)
        if min(dist) == dist_b:
            pred_classes.append(1)
        if min(dist) == dist_c:
            pred_classes.append(2)
            
    return pred_classes
```

```python
# concatenate the datasets a and b 
X = np.concatenate((a, b, c), axis=0)

# perform classification
pred_classes = mahalanobis_classifier_all(X)

# create an array with the true classes of X (= of the datasets a and b)
a_classes = np.zeros(a.shape[0])
b_classes = np.ones(b.shape[0])
c_classes = np.full((c.shape[0]), 2)
true_classes = np.concatenate((a_classes,b_classes,c_classes), axis=None)
```

```python
# create vectors of datapoints for each predicted classes for the plot
a_pred = X[np.where(np.array(pred_classes) == 0)]
b_pred = X[np.where(np.array(pred_classes) == 1)]
c_pred = X[np.where(np.array(pred_classes) == 2)]

# plot classification results
plt.figure(figsize=(12,7))
plot_three_classes(a_pred,b_pred,c_pred,x,y,x_ab,y_ab,Z_ac,Z_bc)
plt.title('Mahalanobis points classification and "complete" Bayes classification decision boundaries')
plt.show()

# compute the number of misclassified samples and the error rate
errors = 0
for i in range(600):
    if (pred_classes[i] != int(true_classes[i])):
        errors += 1

error_rate = (errors/(a.shape[0]+b.shape[0]+c.shape[0]))*100
print('The total number of misclassified samples using Mahalanobis distance is', errors, 'leading to an error rate of {:.2f}%.'.format(error_rate)) 
```

With the 3 classes, the obtained error rate is of 3.5%, with 21 misclassified samples which correspond to error rate of complete bayes classification (multiple binary classification). Again, we see that the complete Bayes classification decision boundaries fit to the obtained classification with Mahalanobis distance, leading to the same conclusions as with only classes `a` and `b`above.

### 2.1 Dataset loading
Here we first declare the methods `extract_data` and `extract_labels` so that we can reuse them later in the code.
Then we extract both the data and corresponding labels, and plot randomly some images and corresponding labels of the training set.

```python
import gzip
import numpy as np
import matplotlib.pyplot as plt

def extract_data(filename, image_shape, image_number):
    with gzip.open(filename) as bytestream:
        bytestream.read(16)
        buf = bytestream.read(np.prod(image_shape) * image_number)
        data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
        data = data.reshape(image_number, image_shape[0], image_shape[1])
    return data


def extract_labels(filename, image_number):
    with gzip.open(filename) as bytestream:
        bytestream.read(8)
        buf = bytestream.read(1 * image_number)
        labels = np.frombuffer(buf, dtype=np.uint8).astype(np.int64)
    return labels
```

### MLP with Pytorch
To gain better insight of how to contruct a MLP, we use the PyTorch library which allows us to create a deep network from scratch.

```python
# data from numpy to tensor 
import torch
from torch import nn
from torch.nn import functional as F
from torch import optim

train_im = train_images.copy()
test_im = test_images.copy()
train_lab = train_labels.copy()
test_lab = test_labels.copy()

train_im = train_im.reshape(train_im.shape[0], 1, train_im.shape[1], train_im.shape[2])
test_im = test_im.reshape(test_im.shape[0], 1, test_im.shape[1], test_im.shape[2])

train_input = torch.from_numpy(train_im)
test_input = torch.from_numpy(test_im)
train_target = torch.from_numpy(train_lab)
test_target = torch.from_numpy(test_lab)

print(train_input.size(), test_input.size(), train_target.size(), test_target.size())
```

We contruct a model that first applies a convolutional filter to the images and then reduce them into a single vector of size 10 corresponding to the class scores (digit 0 to 9). 

Input :&emsp;&emsp; &emsp;&emsp;&emsp;&emsp;&emsp;    60'000 x  1 x 28 x 28 (= images)    <br>
1st convolution :&emsp;&nbsp;&nbsp;&nbsp; 60'000 x 32 x  8 x  8               <br>
2nd convolution :&emsp;&nbsp; 60'000 x 64 x  2 x  2               <br>
view :&emsp;&emsp; &emsp;&emsp;&emsp;&emsp;&emsp;&nbsp;       60'000 x 256                        <br>
MLP layer 1 :&emsp;&emsp;&emsp;&nbsp;&nbsp;   60'000 x 100                        <br>
MLP output :&emsp;&emsp;&emsp;&nbsp;&nbsp;       60'000 x 10 (= class scores)        <br>

We use ReLu as activation function

```python
# The ConvNet module is a simple network that applies a convolution filter before the multi-layer perceptron
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        # Pre-processing with a convolutional filter
        self.conv1 = nn.Conv2d(1, 32, kernel_size=5) 
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5)
        
        # MLP 1 layer, 100 units : The output of the convolutional layer is reduce into a single vector of size 256 -> 100 -> 10 through the MLP
        self.fc1 = nn.Linear(256, 100)
        self.fc2 = nn.Linear(100, 10) # output 10 classes
        
        #Dropout method drop connections between nodes in the MLP during training to reduce training time and avoid over-fitting
        self.dropout = nn.Dropout(0.25)
    
    def forward(self, x):
        
        # Pre-processing with a convolutional filter
        x = F.relu(F.max_pool2d(self.conv1(x), kernel_size=3, stride=3))
        x = F.relu(F.max_pool2d(self.conv2(x), kernel_size=2, stride=2))
       
        # MLP 1 layer, 100 units
        x = self.dropout(F.relu(self.fc1(x.view(-1, 256))))
        x = self.fc2(x)
        
        return x
```

For training, we use SGD as optimizer and Cross Entropy as criterion for loss

```python
# Training 
def train_model(model, train_input, train_target, mini_batch_size): 
        criterion = torch.nn.CrossEntropyLoss()  # Loss criterion = Cross Entropy
        optimizer = optim.SGD(model.parameters(), lr = 1e-3) # Optimizer = SGD
        nb_epochs = 25
        
        for e in range(nb_epochs):
            losses = 0
            for b in range(0, train_input.size(0), mini_batch_size):
                output = model(train_input.narrow(0, b, mini_batch_size)) # prediction
                loss = criterion(output, train_target.narrow(0, b, mini_batch_size)) # computation of the loss
                model.zero_grad()
                loss.backward() # backward pass 
                optimizer.step() # optimizaiton of the weights
                losses += loss.item()
            
            print('Iteration {}, loss = {}'.format(e, losses))

# Computation of errors
def compute_nb_errors(model, data_input, data_target, mini_batch_size):

    nb_data_errors = 0

    for b in range(0, data_input.size(0), mini_batch_size):
        output = model(data_input.narrow(0, b, mini_batch_size))
        _, predicted_classes = torch.max(output, 1)
        for k in range(mini_batch_size):
            if data_target[b + k] != predicted_classes[k]: #if the prediction is not right, increase number of errors
                nb_data_errors = nb_data_errors + 1

    return nb_data_errors
```

Model training and accuracy

```python
mini_batch_size = 500
model = ConvNet()

# Training
train_model(model, train_input, train_target, mini_batch_size)

# Computing number of errors on the test set and train set
nb_train_errors = compute_nb_errors(model, train_input, train_target, mini_batch_size)
nb_test_errors = compute_nb_errors(model, test_input, test_target, mini_batch_size)

# Train and test accuracy
print('train error Net {:0.2f}%'.format(100 - ((100 * nb_train_errors) / train_input.size(0))))
print('test error Net {:0.2f}%'.format(100 - ((100 * nb_test_errors) / test_input.size(0))))
```
